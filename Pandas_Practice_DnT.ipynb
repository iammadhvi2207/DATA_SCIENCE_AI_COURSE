````python
sales_df['Year'] = sales_df['Order_DateTime'].dt.year   # Create a new column for Year
print(sales_df)
````

````python
sales_df['Month_name'] = sales_df['Order_DateTime'].dt.month_name()  
print(sales_df['Month_name'])
````

````python
sales_df['Weekday'] = sales_df['Order_DateTime'].dt.day_name()
print(sales_df)
````
