```python
class Student :
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Roll Number: {self.roll_number}")
        print(f"Name: {self.name}")
        print(f"Marks: {self.marks}")

ob1=Student(1, "John Doe", 85)
ob1.display()
```

```python
class Employee:
    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def display(self):
        print(f"Employee ID: {self.emp_id}")
        print(f"Name: {self.name}")
        print(f"Salary: {self.salary}")

ob2=Employee(101, "Jane Smith", 50000)
ob2.display()
```

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: {amount}. New Balance: {self.balance}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew: {amount}. New Balance: {self.balance}")
        else:
            print("Insufficient funds.")

    def display(self):
        print(f"Account Number: {self.account_number}")
        print(f"Balance: {self.balance}")

ob3=BankAccount("123456789", 1000)
ob3.display()
ob3.deposit(500)
ob3.withdraw(200)
```

```python
n = int(input("Enter a number: "))
if n < 0:
    print("Factorial is not defined for negative numbers.")
elif n == 0:
    print("The factorial of 0 is: 1")
else:
    f=1
    for i in range(1, n+1):
        f *= i
    print(f"The factorial of {n} is: {f}")
```

```python
class Student :
    def __init__(self, name, age, math_grade, science_grade, english_grade):
        self.name = name
        self.age = age
        self.math_grade = math_grade
        self.science_grade = science_grade
        self.english_grade = english_grade

    def calculate_average(self):
        return (self.math_grade + self.science_grade + self.english_grade) / 3
    
ob = Student("Alice", 16, 85, 90, 88)
average = ob.calculate_average()

file = open("Student.txt","w")
file.write(f"Name: {ob.name}\t Age: {ob.age}\n")
file.write(f"Math Grade: {ob.math_grade}\n")
file.write(f"Science Grade: {ob.science_grade}\n")
file.write(f"English Grade: {ob.english_grade}\n")
file.write(f"Average : {average:.2f}\n")
print ("Student details written to Student.txt")
file.close()

    
```
