```python
def Max(n1,n2):
    if(n1>=n2):
        return n1;
    else:
        return n2;
def Min(n1,n2):
    if(n1>=n2):
        return n2;
    else:
        return n1;
def Avg(n1,n2):
    return (n1+n2)/2
def EvenOdd(n1):
    if(n1//2 ==0):
        return "Even"
    else :
        return "Odd"
def Fact(n1):
    if(n1==0):
        return 1
    else:
        return n1*Fact(n1-1)
def FileWrite(s):
    #file=open("Result.txt","w")
    file.write(s + "\n")
    print("Data Written Successfully")

file=open("Result.txt","w")
file.write("Start Of File" + "\n")
y='y'
while (y=='y'):
    n=int(input("Input Choice from 1 to 5 : "))
    if(n==1):
        n1,n2= map(int,input("Enter Any Numbers : ").split())
        print(Max(n1,n2))
        s = f"Maximum of {n1} and {n2} : {Max(n1,n2)}"
        FileWrite(s)
    elif(n==2):
        n1,n2= map(int,input("Enter Any Numbers : ").split())
        print(Min(n1,n2))
        s = f"Minimum of {n1} and {n2} : {Min(n1,n2)}"
        FileWrite(s)
    elif(n==3):
        n1,n2= map(int,input("Enter Any Numbers : ").split())
        print(Avg(n1,n2))
        s = f"Average of {n1} and {n2} : {Avg(n1,n2)}"
        FileWrite(s)
    elif(n==4):
        n1=int(input("Enter Any Number : "))
        print(EvenOdd(n1))
        s = f"{n1} is {EvenOdd(n1)}"
        FileWrite(s)
    elif(n==5):
        n1=int(input("Enter Any Number : "))
        print(Fact(n1))
        s = f"Factorial of {n1} is {Fact(n1)}"
        FileWrite(s)
    else:
        print("Wrong Choice")
    
    y = input("Do You want To Try Again IF YES press 'y' or Press 'n' : ")
file.close()
```

```python
def factorial(n):       #Using Loop
    fact=1
    for i in range(1,n+1,1):
        fact=fact*i
    return fact

num = int(input("Enter a Number : "))
f=factorial(num)
print(f"Factorial of {num} : {f}")
```

```python
def Factorial(n):           # Using Recursion
    if n==0 | n==1:
        return 1
    else:
        return n*Factorial(n-1)
    
num = int(input("Enter a Number : "))
f=Factorial(num)
print(f"Factorial of {num} : {f}")
```

```python
class Mall:
    def __init__(self, name, location, stores):
        self.name = name
        self.location = location
        self.__stores = stores

    def display_info(self):
        print(f"Mall Name: {self.name}")
        print(f"Location: {self.location}")
        print(f"Number of Stores: {self.__stores}")

ob=Mall("Gyan Setu", "Delhi", 50)
ob.display_info()
ob.__stores = 100   # This will not change the private attribute __stores
ob.name = "Ram Ji"
ob.display_info()
```
