```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
import scipy.stats as stat  # this is also a way to import this library package
```

```python


# Student Marks Dataset (out of 100)
marks = np.array([45, 67, 89, 90, 56, 78, 88, 92, 61, 74,
                  83, 95, 52, 68, 71, 85, 77, 64, 58, 81])
marks = pd.DataFrame(marks)
print("Dataset:")
print(marks)
```

```python
print(np.mean(marks))
```

```python
print(np.median(marks))
```

```python
print(stats.mode(marks))
```

```python
print(np.max(marks))
print(np.min(marks))
```

```python
print(np.max(marks)-np.min(marks))
```

```python
print(np.var(marks))
```

```python
print(np.std(marks))
```

```python
marks.describe()
```

```python
print(np.quantile(marks,0.25))
```

```python
print(np.quantile(marks, [0.25, 0.50, 0.75] ))
```

```python
print("IQR : " , (np.quantile(marks, 0.75) - np.quantile(marks, 0.25)))
```

```python
count =(marks > np.mean(marks)).value_counts()
print(count[True])  #wowwwwwww
```

```python
count =(marks < 60).value_counts()
print(count[True])
```

```python
sum = np.sum(marks)
avg = sum / marks.shape[0]
print(sum)
print (avg)
```
