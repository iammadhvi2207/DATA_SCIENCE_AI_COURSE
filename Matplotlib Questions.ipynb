```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Student_ID": [101,102,103,104,105,106,107,108,109,110,
                   111,112,113,114,115,116,117,118,119,120,
                   121,122,123,124,125],

    "Name": ["Amit","Priya","Rahul","Sneha","Karan",
             "Neha","Rohit","Pooja","Vikas","Anjali",
             "Arjun","Meera","Sanjay","Riya","Deepak",
             "Komal","Nitin","Simran","Manish","Kavita",
             "Yash","Isha","Varun","Aarti","Mohit"],

    "Age": [20,21,22,20,23,
            21,22,20,24,23,
            21,22,20,23,24,
            21,22,20,23,24,
            21,22,23,20,21],

    "Course": ["B.Tech","BCA","MCA","B.Sc","B.Tech",
               "MBA","BCA","MCA","B.Tech","MBA",
               "B.Sc","BCA","MCA","MBA","B.Tech",
               "BCA","B.Sc","MBA","MCA","B.Tech",
               "BCA","MBA","B.Sc","MCA","B.Tech"],

    "Marks": [85,78,92,67,88,
              74,81,90,69,95,
              72,84,91,79,87,
              76,89,83,94,68,
              80,77,86,93,82],

    "City": ["Delhi","Mumbai","Lucknow","Jaipur","Pune",
             "Delhi","Noida","Lucknow","Indore","Bhopal",
             "Kanpur","Agra","Patna","Ranchi","Surat",
             "Chennai","Hyderabad","Kolkata","Nagpur","Delhi",
             "Jaipur","Pune","Noida","Lucknow","Mumbai"]
})

# Display DataFrame
print(df)
```

```python
#Create a line chart showing Student_ID vs Marks.
plt.plot(df["Student_ID"],df["Marks"])
```

```python
#Create a line chart of Age vs Marks using: marker,linewidth,linestyle,color,grid,title,legend,Bar Chart
plt.plot(df["Age"],df["Marks"],marker="o",linewidth=1,linestyle="--",color="purple",label="Marks")
plt.grid()
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.legend()

plt.show()
```

```python
#Create a bar chart showing the average marks of each Course.
avg_marks = df.groupby("Course")["Marks"].mean()
plt.bar(avg_marks.index,avg_marks.values,color="purple")
plt.xlabel("Course")
plt.ylabel("Average Marks")
plt.title("Average Marks of Each Course")

plt.show()
```

```python
#Display the number of students in each City using a bar chart.
city_count = df["City"].value_counts()
plt.bar(city_count.index,city_count.values,color="orange")

plt.xlabel("City")
plt.ylabel("Number of Students")
plt.title("Students in Each City")

plt.xticks(rotation=45)

plt.show()
```

```python
#Create a horizontal bar chart showing the top 5 students by Marks.
top5 = df.sort_values(by="Marks", ascending=False).head(5)

plt.figure(figsize=(8,5))

plt.barh(top5["Name"],top5["Marks"],color="purple")

plt.xlabel("Marks")
plt.ylabel("Students")
plt.title("Top 5 Students by Marks")

plt.show()
```

```python
#Create a histogram showing the distribution of Marks using:5 bins,edgecolor,alpha
plt.figure(figsize=(7,5))

plt.hist(df["Marks"],bins=5,edgecolor="black",alpha=0.7)

plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
#Create a histogram of Age with:6 bins,cumulative=True,color='orange',Scatter Plot

plt.hist(df["Age"],bins=6,cumulative=True,color="orange")

plt.xlabel("Age")
plt.ylabel("Cumulative Frequency")
plt.title("Age Distribution")

plt.show()
```

```python
#Create a scatter plot showing Age vs Marks.
plt.scatter(df["Age"],df["Marks"])

plt.xlabel("Age")
plt.ylabel("Marks")
plt.title("Age vs Marks")

plt.show()
```

```python
#Create a scatter plot where:marker size depends on Marks,marker color depends on Age
plt.figure(figsize=(7,5))
plt.scatter(df["Age"], df["Marks"], s=df["Marks"], c=df["Age"])
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
#Create a box plot for Marks.
plt.figure(figsize=(6,5))

plt.boxplot(df["Marks"])

plt.title("Marks Box Plot")
plt.ylabel("Marks")

plt.show()
```

```python
# Sort the DataFrame by Student_ID and create an area chart of Marks.
sorted_df = df.sort_values("Student_ID")

plt.figure(figsize=(8,5))

plt.fill_between(sorted_df["Student_ID"],sorted_df["Marks"],alpha=0.6)

plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.title("Area Chart of Marks")

plt.show()
```
