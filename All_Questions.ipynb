```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sales_data.csv', encoding='utf-8') # utf-8 encoding is used to handle special characters in the CSV file
print(df)
```

```python
# total no. records

print("total records:", df.shape[0]) # [0] is used to get the number of rows in the dataframe

print("\n\n" , df.columns)

# datatype of each column
print("\n\n" ,df.dtypes)
```

```python
# no. of unique products

print("no. of unique products ID: " , df['Product_ID'].nunique()) # nunique() is used to get the number of unique values in a column

print("\nNo. of Unique sales representative: " , df['Sales_Rep'].unique())  # nunique : give total number

# but unique give product name 

print("\nAll unique resigon: " , df['Region'].unique())

print("\n ")
```

```python
# 7. List all Product Categories.

print("\n all Product Categories: " , df['Product_Category'].unique().tolist())

# 8. List all Customer Types.

print("\n list all Customer Type: " , df['Customer_Type'].unique())

# 9. List all Payment Methods.

print("\n list all Payment Mentods: " , df['Payment_Method'].unique())

# 10. List all Sales Channels.

print("\n list all sales Channels: ", df['Sales_Channel'].unique())
```

```python
# 11. Calculate the total Sales Amount.

print("\n Calculate Total Sales Amount: " , df['Sales_Amount'].sum())

# 12. Calculate the average Sales Amount.

print("\n Calculate the Average Sales Amount: " , df['Sales_Amount'].mean())

# 13. Find the maximum Sales Amount.

print("\n Calculate the Maximum Sales Amount: " , df['Sales_Amount'].max())

# 14. Find the minimum Sales Amount.

print("\n Calculate the minimum Sales Amount: " , df['Sales_Amount'].min())

# 15. Calculate the total Quantity sold.

print("\n Calculate Total quantity sold: " , df['Quantity_Sold'].sum())

# 16. Calculate the average Quantity sold.

print("\n Calculate the Average quantity sold: " , df['Quantity_Sold'].mean())

# 17. Calculate the total Unit Cost.

print("\n Calculate Total Unit Cost: " , df['Unit_Cost'].sum())

# 18. Calculate the average Unit Price.

print("\n Calculate the Average Unit Price : " , df['Unit_Price'].mean())

# 19. Calculate the total Discount given.

print("\n Calculate Total Discount Given: " , df['Discount'].sum())

# 20. Calculate the average Discount.

print("\n Calculate the average discount: " , df['Discount'].mean())
```

```python
# 21. Find total Sales Amount by Region.

regi_sal = df.groupby("Region")["Sales_Amount"].sum()
print("Total Sales Amount by Region: " , regi_sal)

# 22. Find average Sales Amount by Region.

avg_sale = df.groupby("Region")["Sales_Amount"].mean()
print("\n Average Sales Amount by Region: " , avg_sale)

# 23. Find total Sales Amount by Sales Representative.

sales_rep = df.groupby("Sales_Rep")["Sales_Amount"].sum()
print("\n Total Sales Amount by Sales Representative: " , sales_rep)

# 24. Find total Quantity sold by Product Category.

pdt_cat = df.groupby("Product_Category")["Quantity_Sold"].sum()
print("\n Total Quantity sold by Product Category: " , pdt_cat)

# 25. Find average Discount by Customer Type.

cust_type = df.groupby("Customer_Type")["Discount"].mean()
print("\n Average Discount by Customer Type: " , cust_type)

# 26. Find total Sales Amount by Payment Method.

pay_meth = df.groupby("Payment_Method")['Sales_Amount'].sum()
print("\n Total Sales Amount by Payment Method: " , pay_meth)

# 27. Find average Sales Amount by Sales Channel.

sales_chan = df.groupby(as_index=False, by="Sales_Channel")["Sales_Amount"].mean()  # as_index = false means 
print("\n Average Sales Amount by Sales Channel: " , sales_chan)

# 28. Find total Sales Amount by Region and Product Category.

ttl_sales = df.groupby(["Region" , "Product_Category"])["Sales_Amount"].sum()
print("\n Total Sales Amount by Region and Product Category: " , ttl_sales)

# 29. Find average Unit Price by Product Category.

avg_unit_price = df.groupby("Product_Category")["Unit_Price"].mean()
print("\n Average Unit Price by Product Category: " , avg_unit_price)

# 30. Find maximum Sales Amount by Sales Representative.

max_sales_rep = df.groupby("Sales_Rep")["Sales_Amount"].max()
print("\n Maximum Sales Amount by Sales Representative: " , max_sales_rep)
```

```python
# 47. Create a new column Profit = (Unit_Price - Unit_Cost) × Quantity.

df['Profit'] = (df['Unit_Price'] - df['Unit_Cost']) * df['Quantity_Sold']

df.info()

```

```python
# 48. Find the Top 10 most profitable transactions.

top_10_profitable = df.nlargest(10, 'Profit')
print("Top 10 most profitable transactions: \n" , top_10_profitable)
```

```python
# 49. Analyze monthly sales trends using Sale_Date.

# Convert date column
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

# Step 2: Extract Year-Month
df["Month"] = df["Sale_Date"].dt.strftime("%Y-%m")

#  Group by Month
monthly_sales = df.groupby("Month")["Sales_Amount"].sum()

# Print Result
print("Monthly Sales Trends: \n", monthly_sales)
```

```python
# Q50. Create an EDA Dashboard containing:
# • Total Sales
# • Total Profit
# • Region-wise Sales
# • Category-wise Sales
# • Monthly Sales Trend
# • Sales by Sales Representative
# • Customer Type Distribution
# • Payment Method Distribution
# • Sales Channel Distribution
# • Discount Analysis

df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]

df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

monthly_sales = df.groupby(df["Sale_Date"].dt.to_period("M"))["Sales_Amount"].sum()

plt.figure(figsize=(18, 12))

plt.subplot(3,3,1)
df.groupby("Region")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Region-wise Sales")

plt.subplot(3,3,2)
df.groupby("Product_Category")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Category-wise Sales")

plt.subplot(3,3,3)
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")

plt.subplot(3,3,4)
df.groupby("Sales_Rep")["Sales_Amount"].sum().plot(kind="bar")
plt.title("Sales by Sales Representative")

plt.subplot(3,3,5)
df["Customer_Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Customer Type Distribution")
plt.ylabel("")

plt.subplot(3,3,6)
df["Payment_Method"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.ylabel("")

plt.subplot(3,3,7)
df["Sales_Channel"].value_counts().plot(kind="bar",color='orange')
plt.title("Sales Channel Distribution")

plt.subplot(3,3,8)
plt.hist(df["Discount"], bins=10,color='green')
plt.title("Discount Analysis")

plt.subplot(3,3,9)
plt.hist(df["Profit"], bins=10,color='darkblue')
plt.title("Profit Distribution")

plt.tight_layout()
plt.show()

print("Total Sales:", df["Sales_Amount"].sum())
print("Total Profit:", df["Profit"].sum())
```
