```python
import pandas as pd
list1 = [1, 2, 3, 4, 5]
list2 = ['A', 'B', 'C', 'D', 'E']

df = pd.Series(list1)  # Create a Series from a list
print(df)  # Display the Series
df = pd.DataFrame({'Numbers': list1, 'Letters': list2})  # Create a DataFrame from two lists
print(df)  # Display the DataFrame
```
