```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Student_ID": [101,102,103,104,105,106,107,108,109,110,
                   111,112,113,114,115,116,117,118,119,120],

    "Name": ["Aman","Riya",None,"Karan","Priya","Rahul","Neha",None,
             "Vikas","Pooja","Ankit","Simran","Rohit",None,"Kajal",
             "Deepak","Meena","Arjun","Nisha",None],

    "Age": [20,21,np.nan,22,19,23,21,20,np.nan,22,
            24,23,21,20,22,np.nan,19,21,23,20],

    "City": ["Delhi","Mumbai","Pune",None,"Jaipur","Delhi","Mumbai",
             "Pune","Delhi",None,"Jaipur","Mumbai","Delhi","Pune",
             None,"Delhi","Mumbai","Jaipur","Pune","Delhi"],

    "Marks": [85,90,78,np.nan,88,76,91,82,95,np.nan,
              80,84,79,92,87,75,np.nan,89,81,93]
})

print(df)
```

```python
df.isnull()  # Check for missing values in the DataFrame
```

```python
df.dropna()  # Drop rows with any missing values from the DataFrame
print(df)
```

```python
df.fillna(0)  # Fill missing values with 0
print(df)
```

```python
df.fillna(value={'Name': 'Unknown', 'City': 'Unknown', 'Age': 0, 'Marks': 0}, inplace=True)  # Fill missing values with specified values in place
print(df)
```

```python
mean_marks = df['Marks'].mean()  # Calculate the mean of the 'Marks' column
print(mean_marks)
```

```python
df.fillna(value = {'Name': 'Unknown', 'City': 'Unknown', 'Age': 0, 'Marks': mean_marks}, inplace=True)  # Fill missing values with specified values in place
df['Marks'].replace(to_replace=np.nan, value=mean_marks, inplace=True)  # Replace missing values in 'Marks' with the mean
print(df)
```
