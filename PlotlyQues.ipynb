```python
import pandas as pd

df = pd.DataFrame({
    "Order_ID": range(101, 121),
    "Product": [
        "Laptop","Mobile","Tablet","Laptop","Mobile",
        "Tablet","Laptop","Mobile","Tablet","Laptop",
        "Mobile","Tablet","Laptop","Mobile","Tablet",
        "Laptop","Mobile","Tablet","Laptop","Mobile"
    ],
    "Category": [
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics"
    ],
    "City": [
        "Delhi","Mumbai","Lucknow","Delhi","Lucknow",
        "Mumbai","Delhi","Lucknow","Mumbai","Delhi",
        "Lucknow","Mumbai","Delhi","Lucknow","Mumbai",
        "Delhi","Lucknow","Mumbai","Delhi","Lucknow"
    ],
    "Sales": [
        65000,30000,25000,70000,32000,
        27000,68000,35000,26000,72000,
        33000,28000,75000,34000,29000,
        76000,36000,30000,78000,37000
    ],
    "Profit": [
        12000,5000,4500,13000,5500,
        4700,12500,6000,4800,14000,
        6200,5000,14500,6300,5200,
        15000,6500,5400,15500,6800
    ],
    "Quantity": [
        5,10,8,6,9,
        7,5,11,8,6,
        10,7,5,9,8,
        4,12,9,5,10
    ]
})

print(df)
```

```python
import plotly.express as px

fig= px.bar(df, x='Product', y='Sales', color='Product', title='Total Sales by Product')
fig.show()
```

```python
fig = px.pie(df,names='Product',values='Sales',color='Product',title='Total sales')
fig.show()
```

```python
fig= px.scatter(
    df,
    x='Sales',
    y='Profit',
    size= 'Quantity',
    color= 'Product'
)
fig.show()
```

```python
fig = px.histogram(
    df,
    x='Sales',
    nbins=10,
    title='Histogram of Sales'
)
fig.show()
```

```python
fig = px.box(
    df,
    x='Product',
    y='Sales',
    title='Sales Distributiom'
)
fig.show()
```

```python
sales_by_city = df.groupby("City")["Sales"].sum().reset_index()

fig = px.bar(
    sales_by_city,
    x='City',
    y='Sales',
    color='City',
    title= 'CItywide Sales'
)
fig.show()
```

```python
profit_product = df.groupby('Product')['Profit'].mean().reset_index()

fig = px.line(
    profit_product,
    x='Product',
    y='Profit',
    title='Mean Profit'
)
fig.show()
```
