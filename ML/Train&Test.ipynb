## Logistic Regression

(All Data Given Should Be in Numeric Form)

If the data is not Numeric Convert it into numeric data

```python
import pandas as pd


data = {
    "StudyHours" : [1,2,3,4,5,6,7,8],
    "Result" : ["FAIL", "FAIL", "FAIL", "PASS", "PASS", "PASS", "PASS", "PASS"]
}

df = pd.DataFrame(data)

df["Result"] = df["Result"].map({"FAIL":0 , "PASS" : 1})    # Changing String value to Numeric value
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder    # using this also we can convert data to Numeric Values

data = {
    "StudyHours" : [1,2,3,4,5,6,7,8],
    "Result" : ["FAIL", "FAIL", "FAIL", "PASS", "PASS", "PASS", "PASS", "PASS"]
}

df = pd.DataFrame(data)
le= LabelEncoder()
df["Result"]=le.fit_transform(df["Result"])
print(df)
```

```python
from sklearn.model_selection import train_test_split        # to divide Data for training and testing
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score                  # This Library is to calculate the accuracy of the model

X = df[["StudyHours"]]
Y= df["Result"]

X_train , X_test , Y_train , Y_test = train_test_split(
    X,
    Y,
    test_size=0.25,
    random_state=42
)

model= LogisticRegression()
model.fit(X_train, Y_train)
pred = model.predict(X_test)
accuracy = accuracy_score(Y_test,pred)
print(accuracy)
print(pred)
print(Y_test)
print(X_test)
```
