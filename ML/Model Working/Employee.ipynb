```python
df.fillna(value = {'Education' : 'Unknown' , 'City' : 'Unknown', 'Salary' : 0}, inplace=True)
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

city_sal = df.groupby('City')['Salary'].sum()
plt.bar(city_sal.index,city_sal.values, color = 'green')
plt.xlabel('City')
plt.ylabel('Salary')
plt.title('City wise Salary')
plt.show()
```

```python
gender_diff = df.groupby(['City', 'Gender']).size().unstack()

gender_diff.plot(kind='bar', figsize=(10, 5), color=['blue', 'green'])

plt.xlabel('City')
plt.ylabel('Employee Count')
plt.title('Employee Count by City and Gender')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()
```
