```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Age":[22,23,24,25,26,27,28,30,
           45,47,50,52,55,58,60],
    "Income":[90000,85000,95000,87000,92000,
              89000,91000,88000,
              40000,42000,38000,
              36000,35000,30000,28000]
})

X = df[["Age","Income"]]

model = KMeans(
    n_clusters=2,
    random_state=42
)

df["Cluster"] = model.fit_predict(X)

print(df)
```
