```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans


df = pd.read_csv('customer_segmentation.csv' )
print(df.head())
```

```python
X = df[["Age" , "Annual_Income" , "Spending_Score"]]

model = KMeans(n_clusters = 3 , random_state = 42)

df['Cluster'] = model.fit_predict(X)

print(df)
```
