```python
import numpy as np

# Matrix A
A = np.array([
    [2, 4, 6],
    [8, 10, 12],
    [14, 16, 18]
])

# Matrix B
B = np.array([
    [1, 3, 5],
    [7, 9, 11],
    [13, 15, 17]
])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
```

```python
print(A.shape)
print(B.shape)
print(A.size)
print(B.size)
print(np.ndim(A), " ", A.ndim)
print(np.ndim(B))
```

```python
print(A+B)
print(A-B)
print(A*B)
print(np.matmul(A,B))
print(A.T)
print(B.T)
```

```python
print(np.linalg.det(A))
```

```python
print(np.linalg.inv(A))
```

```python
print(np.trace(A))  # Trace GIves the Sum of Diagonal elements
print(np.trace(B))
```

```python
print(A.max())
print(A.min())
print(A.mean())
print(A.sum())
```

```python
print(B.sum(axis=1))    # Row Wise
print(B.sum(axis=0))    # Column Wise
```

```python
print(A[0])
print(A[:,1])
print(A[-1])
print(A[-1][-1])
```

```python
print(A.reshape(1,9))
print(B.flatten())
```

```python
print(np.linalg.eig(A))
print(np.linalg.eig(B))
```

```python
print(A)
print(B)

print(np.hstack((A,B)))
print(np.vstack((A,B)))
```
