##PROJECT 1: STUDENT MARKS MANAGEMENT SYSTEM (ARRAY)

1.Ôüá ÔüáCreate an array to store marks of 10 students.

2.Ôüá ÔüáInput marks from the user.

ÔÇç3.Ôüá ÔüáDisplay all marks.

ÔÇç4.Ôüá ÔüáFind the total marks.

ÔÇç5.Ôüá ÔüáCalculate the average marks.

ÔÇç6.Ôüá ÔüáFind the highest mark.

ÔÇç7.Ôüá ÔüáFind the lowest mark.

ÔÇç8.Ôüá ÔüáCount students who scored above 75.

ÔÇç9.Ôüá ÔüáCount students who failed (marks < 40).

10.Ôüá ÔüáSearch for a particular mark.

11.Ôüá ÔüáSort marks in ascending order.

12.Ôüá ÔüáSort marks in descending order.

13.Ôüá ÔüáDisplay only even marks.

14.Ôüá ÔüáReplace a specific mark with a New value

```python
import numpy as np
marks = np.zeros(10)
for i in range(10):
    marks[i] = float(input(f"Enter marks for student {i+1}: "))

print("Marks of students:", marks)
print("Sum of marks:", np.sum(marks))
print("Average marks:", np.mean(marks))
print("Highest mark:", np.max(marks))
print("Lowest mark:", np.min(marks))

count_passed = np.sum(marks > 75)
print(f"Number of students who passed (marks > 75): {count_passed}")

count_failed = np.sum(marks < 40)
print("Number of students who failed (marks < 40):", count_failed)

mark = float(input("Enter a mark to check if any student has it: "))
if mark in marks:
    for i in range(len(marks)):
        if marks[i] == mark:
            print(f"Student {i+1} has the mark {mark}.")
            break

else :
    print(f"No student has the mark {mark}.")

sorted_marks = np.sort(marks)
print("Marks in ascending order:", sorted_marks)
print("Marks in descending order:", sorted_marks[::-1])

for i in range(len(marks)):
    if marks[i] % 2 ==0 :
        print(marks[i], "  ")
print("\n")

replace_mark = float(input("Enter the mark you want to replace: "))
new_marks = float(input("Enter the new mark: "))

if replace_mark not in marks:
    print(f"No student has the mark {replace_mark}.")
else:
    for i in range(len(marks)):
        if marks[i] == replace_mark:
            marks[i] = new_marks
            print(f"Replaced mark {replace_mark} with {new_marks} for student {i+1}.")
            break
```
