##PROJECT 1: STUDENT MARKS MANAGEMENT SYSTEM (ARRAY)

1.вЃ  вЃ Create an array to store marks of 10 students.

2.вЃ  вЃ Input marks from the user.

вЂ‡3.вЃ  вЃ Display all marks.

вЂ‡4.вЃ  вЃ Find the total marks.

вЂ‡5.вЃ  вЃ Calculate the average marks.

вЂ‡6.вЃ  вЃ Find the highest mark.

вЂ‡7.вЃ  вЃ Find the lowest mark.

вЂ‡8.вЃ  вЃ Count students who scored above 75.

вЂ‡9.вЃ  вЃ Count students who failed (marks < 40).

10.вЃ  вЃ Search for a particular mark.

11.вЃ  вЃ Sort marks in ascending order.

12.вЃ  вЃ Sort marks in descending order.

13.вЃ  вЃ Display only even marks.

14.вЃ  вЃ Replace a specific mark with a New value

```python
import numpy as np
marks = np.zeros(10)
for i in range(10):
    marks[i] = float(input(f"Enter marks for student {i+1}: "))

print("Marks of students:", marks)
print("Sum of marks:", np.sum(marks))
print("Average marks:", np.mean(marks))
print("Highest mark:", np.max(marks))
print("Lowest mark:", np.min(marks))

count_passed = np.sum(marks > 75)
print(f"Number of students who passed (marks > 75): {count_passed}")

count_failed = np.sum(marks < 40)
print("Number of students who failed (marks < 40):", count_failed)

mark = float(input("Enter a mark to check if any student has it: "))
if mark in marks:
    for i in range(len(marks)):
        if marks[i] == mark:
            print(f"Student {i+1} has the mark {mark}.")
            break

else :
    print(f"No student has the mark {mark}.")

sorted_marks = np.sort(marks)
print("Marks in ascending order:", sorted_marks)
print("Marks in descending order:", sorted_marks[::-1])

for i in range(len(marks)):
    if marks[i] % 2 ==0 :
        print(marks[i], "  ")
print("\n")

replace_mark = float(input("Enter the mark you want to replace: "))
new_marks = float(input("Enter the new mark: "))

if replace_mark not in marks:
    print(f"No student has the mark {replace_mark}.")
else:
    for i in range(len(marks)):
        if marks[i] == replace_mark:
            marks[i] = new_marks
            print(f"Replaced mark {replace_mark} with {new_marks} for student {i+1}.")
            break
```

Hospital Management System

Develop a Hospital Management System using Multilevel Inheritance.

Requirements:

Create a class Hospital.

Constructor should initialize:

Hospital Name

City

Create a class Doctor inheriting from Hospital.

Constructor should initialize:

Doctor Name

Specialization

Consultation Fee

Call the parent constructor.

Create a class Patient inheriting from Doctor.

Constructor should initialize:

Patient Name

Age

Disease

Number of Consultation Days

Call the parent constructor.

Display:

Hospital Information

Doctor Information

Patient Information

Total Consultation Charges

GST (18%)

Final Bill

```python
class Hospital :
    def __init__ (self, hos_name, city):
        self.hos_name = hos_name
        self.city = city

    def display_hospital_info(self):
        print(f"Hospital Name: {self.hos_name}")
        print(f"City: {self.city}")

class Doctor(Hospital):
    def __init__ (self, name, specialization, fee, hos_name, city):
        self.name=name
        self.specialization= specialization
        self.fee=fee
        super().__init__(hos_name, city)

    def display_doctor_info(self):
        
        print(f"Doctor Name: {self.name}")
        print(f"Specialization: {self.specialization}")
        print(f"Consultation Fee: ${self.fee}")

class Patient(Doctor):
    def __init__(self, p_name, age, disease, days, name, specialization, fee, hos_name, city):
        self.p_name=p_name
        self.age=age
        self.disease=disease
        self.days=days
        super().__init__(name, specialization, fee, hos_name, city)

    def display_patient_info(self):
        super().display_hospital_info()
        super().display_doctor_info()
        print(f"Patient Name: {self.p_name}")
        print(f"Age: {self.age}")
        print(f"Disease: {self.disease}")
        print(f"Days Admitted: {self.days}")

    def calculate_bill(self):
        total_bill = self.fee * self.days
        print(f"Total Bill for {self.p_name}: ${total_bill}")
        print(f"GST : ${total_bill * 0.18}")
        print(f"Final Bill : ${total_bill + (total_bill * 0.18)}")

patient1=Patient("John Doe", 30, "Flu", 3, "Dr. Smith", "General Physician", 100, "City Hospital", "New York")
patient1.display_patient_info()
patient1.calculate_bill()
```
