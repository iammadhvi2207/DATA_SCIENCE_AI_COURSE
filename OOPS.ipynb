```python
class Student :
    course ="Data Science"
    def __init__(self,rn,n,c,p,ds):
        self.rollno=rn
        self.name=n
        self.clan=c
        self.python=p
        self.datascience=ds
        
    def displayStudentReport(self):
        total=self.clan+self.python+self.datascience
        per=total/300*100
        grade=""
        if per>=80:
            grade="A"
        elif per>=60:
            grade="B"
        elif per>=40:
            grade="C"
        else:
            grade="D"

        print("\n---------------------------------STUDENT REPORT CARD-----------------------------------")
        print("-----------------------------------------------------------------------------------------")
        print(f"Roll No.: {self.rollno}          Name: {self.name}")
        print("------------------------------------------------------------------------------------------")
        print(f"CLanguage: {self.clan}")
        print(f"Python: {self.python}")
        print(f"Data Science: {self.datascience}")
        print("------------------------------------------------------------------------------------------")
        print(f"Total: {total}       Percentage: {per}%           Grade: {grade}")
        print("------------------------------------------------------------------------------------------")
        
s1 = Student(1 , "Kriti" , 90 , 80 , 70 )
s2 = Student(2 , "Rohit" , 30 , 60 , 50)
s1.displayStudentReport()
s2.displayStudentReport()
```

```python
def add(n1,n2):     # Overloading Not Supported in Python, but we can achieve it using default arguments or variable-length arguments.
    return n1+n2

def add(n1,n2,n3):
    return n1+n2+n3

n1,n2,n3 = map(int,input("Enter three numbers separated by space: ").split())
result = add(n1,n2,n3)
print(f"The sum of {n1}, {n2}, and {n3} is: {result}")
result = add(n1,n2)
print(f"The sum of {n1}, {n2}, and {n3} is: {result}")
```

```python
class Account:
    def __init__(self, user, balance, pin):
        self.user = user          # Public: Access from anywhere
        self._balance = balance   # Protected: Convention for class/subclass
        self.__pin = pin          # Private: Triggers name mangling

# Instantiate the object
acc = Account("Alice", 1500, 9999)

# 1. Accessing Public
print(acc.user)        # Output: Alice

# 2. Accessing Protected (Works, but breaks convention)
print(acc._balance)    # Output: 1500

# 3. Accessing Private
#print(acc.__pin)     # Throws AttributeError!

# Bypassing Private restriction via Name Mangling
print(acc._Account__pin) # Output: 9999
```

```python
# Parent Class
class Vehicle:
    def __init__(self, brand, plate_number):
        self.__brand = brand
        self._plate_number = plate_number

    def move(self):
        return f"The {self.__brand} with plate number {self._plate_number} is moving."

# Child Class
class Car(Vehicle):
    def honk(self):
        return "Beep beep!"

# Usage
my_car = Car("Toyota","4567H")
print(my_car.move())  # Inherited method: Output -> The Toyota is moving.
print(my_car.honk())  # Child method: Output -> Beep beep!
```
